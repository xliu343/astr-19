```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
shape = 2.0
scale = 1.0
random_numbers = np.random.gamma(shape, scale, 1000)
fig, ax = plt.subplots()
ax.hist(random_numbers, bins=100)
ax.set_xlabel("Random Number Value (Gamma Distribution)")
ax.set_ylabel("Count")
ax.set_title("Histogram of Gamma-Distributed Random Numbers")
fig.savefig("gamma_distribution_histogram.pdf")
plt.show()
```
