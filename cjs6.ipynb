# Trigonometric Functions Notebook

This notebook defines functions for `sin(x)` and `cos(x)`, tabulates their values over a range of `x`,
and prints selected values in a clear, column-based format.

```python
import math

def sin_func(x):
    """
    Returns the sine of x (x in radians).
    """
    return math.sin(x)
```

## sin(x) Function

This function takes a numerical value `x` (in radians) and returns the sine of `x`
using Python's `math.sin()` function.

```python
def cos_func(x):
    """
    Returns the cosine of x (x in radians).
    """
    return math.cos(x)
```

## cos(x) Function

This function takes a numerical value `x` (in radians) and returns the cosine of `x`
using Python's `math.cos()` function.

```python
import numpy as np

x_values = np.linspace(0, 2, 1000)

sin_values = [sin_func(x) for x in x_values]
cos_values = [cos_func(x) for x in x_values]
```

## Tabulating sin(x) and cos(x)

Here we generate 1000 evenly spaced values of `x` between 0 and 2,
and compute `sin(x)` and `cos(x)` using the previously defined functions.

```python
print("x\t\tsin(x)\t\tcos(x)")
print("--------------------------------")

for i in range(10):
    print(f"{x_values[i]:.6f}\t{sin_values[i]:.6f}\t{cos_values[i]:.6f}")
```
