```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 1, 100)
def sin_function(x):
    return np.sin(x)
def cos_function(x):
    return np.cos(x)
```

```python
y_sin = sin_function(x)
y_cos = cos_function(x)
fig, ax = plt.subplots(1, 2)
ax[0].plot(x, y_sin)
ax[0].set_title("sin(x)")
ax[0].set_xlabel("x")
ax[0].set_ylabel("sin(x)")
ax[1].plot(x, y_cos)
ax[1].set_title("cos(x)")
ax[1].set_xlabel("x")
ax[1].set_ylabel("cos(x)")
plt.tight_layout()
fig.savefig("sin_cos_mp.pdf")
plt.show()
```
